```python
from citipy import citipy
```

```python
# Create a empty list to append the cities
cities = []

# To collect a certain amount of cities
while len(cities)<643:

    # Randomly choose the cities from a certain range
    x = (random.choice(range(-90, 90)))
    y = (random.choice(range(-180,180)))
    
    # Collect the cities names nearest to the range I set up for x and y
    city = citipy.nearest_city(x,y).city_name
    
    # If the city is not in the empty cities list then to append it
    if city not in cities:
        cities.append(city)
# Count of cities        
len(cities)
```

```python
# Create url from the OpenWeather map
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&appid=" + api_key
```

```python
# Create emtpy list to collect the data needed to create the dataframe
city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

# Define the count of pass through
count = 1
sets = 1

print(f"Beginning Data Retrieval")
print(f"-----------------------------")

# Loop through the city
for city in cities:
    
    # Setting if statement of when to create a new set and adding 1 to set
    if count > 50:
        sets += 1
        count = 0
    
    # Creating a print log of each city as it's being processed
    print(f'Processing Record %s of Set %s | %s' % (count, sets, city))
    
    # Adding 1 to count
    count += 1
    
    # Run the code
    try:
        
        # Use the requests.get and .json on the url
        response = requests.get(f"{url}&q={city}").json()
        
        # Append to the empty lists I created above
        city_name.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        
    # When it doesn't find a city, it will print the statement and continue searching     
    except:
        print("City not found. Skipping...")
    continue

print(f"--------------")
print(f"Data Retrieval Complete")
print(f"--------------")
```

```python
# Create a dictionary of the data I collected from the empty list above
weather_dictionary = {
            "City": city_name,
            "Lat": lat,
            "Lng": lng,
            "Max Temp": max_temp,
            "Humidity": humidity,
            "Cloudiness": cloudiness,
            "Wind Speed": wind_speed,
            "Country": country,
            "Date": date,
}

# Set it up as a DataFrame
weather_city_data = pd.DataFrame(weather_dictionary)

# Export the data into a .csv file
weather_city_data.to_csv('CSV Data/weather_city_data.csv',index=False)

# Use count on the DataFrame to see if the data equals to each other and how much data we have
weather_city_data.count()
```

```python
# Split the DataFrame into a Northern Hemisphere and Southern Hemisphere
northern_hemisphere = weather_city_data.loc[weather_city_data["Lat"] >= 0]
southern_hemisphere = weather_city_data.loc[weather_city_data["Lat"] < 0]
```

```python
# Northern Hemisphere - Max Temp vs. Latitude Linear Regression  
# Set up the linear regression
x_values = northern_hemisphere['Lat']
y_values = northern_hemisphere['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
print(f"The r-value is: {rvalue}")
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Plot the scatter graph and linear regression
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(0,50), fontsize=15,color="red")

plt.ylim(0, 100)
plt.xlim(0, 80)

plt.title("Northern Hemisphere - Max Temp vs. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Max Temp")
plt.show()
```

```python
# Southern Hemisphere - Max Temp vs. Latitude Linear Regression
# Set up the linear regression
x_values = southern_hemisphere['Lat']
y_values = southern_hemisphere['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
print(f"The r-value is: {rvalue}")
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Plot the scatter graph and linear regression
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(0,50), fontsize=15,color="red")

plt.ylim(38, 105)
plt.xlim(-60, 5)

plt.title("Southern Hemisphere - Max Temp vs. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Max Temp")
plt.show()
```
